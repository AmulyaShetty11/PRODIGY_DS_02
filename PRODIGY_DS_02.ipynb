## Handling Missing Values

```python
data.isnull().sum()
```

```python
data["Age"].fillna( data["Age"].mean() , inplace = True)
data["Cabin"].fillna("N/A" , inplace = True)
data.dropna(subset=["Embarked"], inplace=True)
data.isnull().sum()
```

```python
data.duplicated().sum()
```

## Data Visualization

```python
# histogram
plt.title("Age Distribution")
sns.histplot(x='Age', data=data, kde=True )
plt.show()
```

```python
plt.figure(figsize=(11, 5))
plt.title("Scatter Plot of Age vs Fare vs Sex")
sns.scatterplot(data=data, x="Age", y="Fare", hue="Survived", size="Sex")
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()    
```

```python
sns.pairplot(data, hue='Sex', height=2)
```
